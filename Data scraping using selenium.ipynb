```python
products=browser.find_elements_by_css_selector('li.product-base')
len(products)
```

```python
for item in products:
    print(item)
```

```python
for item in products:
    name=item.find_element_by_css_selector('div.product-brand')
    name=name.get_attribute('innerHTML')
    desc=item.find_element_by_css_selector('h4.product-product')
    desc=desc.get_attribute('innerHTML')
    discount=item.find_element_by_css_selector('span.product-discountPercentage')
    discount=discount.get_attribute('innerHTML')
    print(name ,'|', desc,' | ',discount)
```

```python
df_products=pd.DataFrame(columns=['name','description','price','discount'])
for item in products:
    name=item.find_element_by_css_selector('div.product-brand')
    name=name.get_attribute('innerHTML')
    desc=item.find_element_by_css_selector('h4.product-product')
    desc=desc.get_attribute('innerHTML')
    discount=item.find_element_by_css_selector('span.product-discountPercentage')
    discount=discount.get_attribute('innerHTML')
    parent_price=item.find_element_by_css_selector('div.product-price')
    child_price=parent_price.find_element_by_css_selector('span')
    #parent_price=print(parent_price.get_attribute('innerHTML'))
    try:
        price=parent_price.find_element_by_css_selector('span.product-discountedPrice')
        price=price.get_attribute('innerHTML')
    except:
        price=child_price.get_attribute('innerHTML')
    curr={'name':name,
         'description':desc,
         'price':price,
         'discount':discount}
    df_products=df_products.append(curr,ignore_index=True)
df_products
```

# NAUKRI WEBSITE

```python
browser=webdriver.Chrome('chromedriver.exe')
browser.get('https://www.naukri.com/data-scientist-jobs')
```

```python
#find repeating line...
#1. identify the parent tag
#whenver we want to extract

jobs=browser.find_elements_by_css_selector('div.row')
print(len(jobs))
```

```python
jobs=browser.find_elements_by_css_selector('div.row[type="tuple"]')
print(len(jobs))
```

```python
for job in jobs:
    title=job.find_element_by_css_selector('li.desig')
    title=title.get_attribute('title')
    
    company=job.find_element_by_css_selector('span.org')
    company=company.get_attribute('innerHTML')
    
    salary=job.find_element_by_css_selector('span.salary')
    salary=salary.get_attribute('innerHTML')
    
#the next block contains the entire code to find the jobs with its attributes    
```

```python
#the above code in terms of a function

job_info=pd.DataFrame(columns=['title','company','salary','date','location','experience','skill','description'])
def get_info(job,selector,attribute):
    try:
        tag=job.find_element_by_css_selector(selector)
        tag=tag.get_attribute(attribute)
    except:
        tag=None
    return tag

for job in jobs:
    title=get_info(job,'li.desig','title')
    company=get_info(job,'span.org','innerHTML')
    salary=get_info(job,'span.salary','innerHTML')
    date=get_info(job,'span.date','innerHTML')
    
    parent_location=job.find_element_by_css_selector('span.loc')
    location=get_info(parent_location,'span','innerHTML')
    
    experience=get_info(job,'span.exp','innerHTML')
    skill=get_info(job,'span.skill','innerHTML')
    description=get_info(job,'span.desc','innerHTML')
    curr_job={'title':title,'company':company,'salary':salary,'date':date,
              'location':location,'experience':experience,'skill':skill,'description':description}
    job_info=job_info.append(curr_job,ignore_index=True)
job_info
    
```

```python
def get_salary(value,type_salary):
    if type_salary=='low':
        try:
            return float(value.split('-')[0])
        except:
            return 0
    else:
        try:
            return float(value.split('-')[1])
        except:
            return 0
```

```python
naukri['low_sal']=naukri['sal_reg'].apply(lambda x: get_salary(x,'low'))
naukri['high_sal']=naukri['sal_reg'].apply(lambda x: get_salary(x,'high'))
naukri.head()
```
