```python
book_info=pd.DataFrame(columns=['category','book name','author name','description','votes',
                                'total votes','image url','book link url'])
for book in books:
    
    category=book.find_element_by_css_selector('h4')
    category=category.get_attribute('innerHTML')
    
    book_name=book.find_element_by_css_selector('img')
    book_name=book_name.get_attribute('alt')
    
    book_img_url=book.find_element_by_css_selector('img')
    book_img_url=book_img_url.get_attribute('src')
    
    book_url=book.find_element_by_css_selector('a')
    book_url=book_url.get_attribute('href')
    
    curr_a={'category':category,'book name':book_name,'image url':book_img_url,
          'book link url':book_url}
    book_info=book_info.append(curr_a,ignore_index=True)
book_info 

        
    
    
    
    
    
    
```

```python
book_temp=pd.DataFrame(columns=['author name','description','votes','total votes'])
for url in book_info['book link url']:
    browser.get(url)
    book2=browser.find_elements_by_css_selector('div.pollContents')
    for book1 in book2:
        author_name=book1.find_element_by_css_selector('a.authorName')
        author_name=author_name.find_element_by_css_selector('span')
        author_name=author_name.get_attribute('innerHTML')
        
        description=book1.find_element_by_css_selector('div.readable.stacked.gcaBookDescription')
        description=description.get_attribute('innerHTML')
        
        votes=book1.find_element_by_css_selector('span.greyText.gcaNumVotes')
        votes=votes.get_attribute('innerHTML')
        
        total_votes=book1.find_element_by_css_selector('h2.gcaSubheader')
        total_votes=total_votes.find_element_by_css_selector('greytext.gcaNumVotes')
        total_votes=total_votes.get_attribute('innerHTML')
        
        
        curr_b={'author name': author_name,'description':description,
               'votes':votes,'total votes':total_votes}
    book_temp=book_temp.append(curr_b,ignore_index=True)
book_temp
        
    
```

```python
def number_alone(value):
        number=value.str.replace('[^0-9]')
        return number

book_temp['votes']=book_temp['votes'].apply(lambda x:number_alone(x))
```

```python
book_info=pd.DataFrame(columns=['category','book name','author name','description','votes',
                                'total votes','image url','book link url'])
for book in books:
    
    category=book.find_element_by_css_selector('h4')
    category=category.get_attribute('innerHTML')
    
    book_name=book.find_element_by_css_selector('img')
    book_name=book_name.get_attribute('alt')
    
    book_img_url=book.find_element_by_css_selector('img')
    book_img_url=book_img_url.get_attribute('src')
    
    book_url=book.find_element_by_css_selector('a')
    book_url=book_url.get_attribute('href')

    
    for url in book_info['book link url']:
        browser.get(url)
        book2=browser.find_elements_by_css_selector('div.pollContents')
        for book1 in book2:
            author_name=book1.find_element_by_css_selector('a.authorName')
            author_name=author_name.find_element_by_css_selector('span')
            author_name=author_name.get_attribute('innerHTML')
        
            description=book1.find_element_by_css_selector('div.readable.stacked.gcaBookDescription')
            description=description.get_attribute('innerHTML')
        
            votes=book1.find_element_by_css_selector('span.greyText.gcaNumVotes')
            votes=votes.get_attribute('innerHTML')
        
            total_votes=book1.find_element_by_css_selector('h2.gcaSubheader')
            total_votes=total_votes.find_element_by_css_selector('greytext.gcaNumVotes')
            total_votes=total_votes.get_attribute('innerHTML')
        
            currb={'category':category,'book name':book_name,'image url':book_img_url,
                    'book link url':book_url,'author name': author_name,'description':description,
                    'votes':votes,'total votes':total_votes}
            book_info=book_info.append(currb,ignore_index=True)
book_info
        
```
