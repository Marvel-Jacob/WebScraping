```python
import selenium
import pandas as pd
from selenium import webdriver
```

```python
browser=webdriver.Chrome('C:/Users/Administrator/Python DEMOS/DEMOS/chromedriver.exe')
```

```python
browser.get('https://www.gsmarena.com/apple_iphone_6-reviews-6378.php')
```

```python
reviews=browser.find_elements_by_css_selector('div.user-thread')
print(reviews)
```

```python
review_info=pd.DataFrame(columns=['user name','review date','review text','up votes','rate value'])
for review in reviews:
    user_name=review.find_element_by_css_selector('ul.uinfo2')
    user_name=user_name.find_element_by_css_selector('li')
    user_name=user_name.get_attribute('innerHTML')
    
    reviewdate=review.find_element_by_css_selector('li.upost')
    reviewdate=reviewdate.find_element_by_css_selector('time')
    reviewdate=reviewdate.get_attribute('innerHTML')
    
    reviewtext=review.find_element_by_css_selector('p.uopin')
    reviewtext=reviewtext.get_attribute('innerHTML')
    
    try:
        upvotes=review.find_element_by_css_selector('span.uavatar-stat.upvote-count')
        upvotes=upvotes.get_attribute('innerHTML')
    except:
        upvotes=None
    
    ratevalue=review.find_element_by_css_selector('span.thumbs-score')
    ratevalue=ratevalue.get_attribute('innerHTML')
    
    rev={'user name': user_name,'review date':reviewdate,
               'review text':reviewtext,'up votes':upvotes,'rate value':ratevalue}
    review_info=review_info.append(rev,ignore_index=True)
review_info
    
##nextpage=review.find_element_by_css_selector('div#user-pages.nav-pages')
##nextpage=nextpage.find_element_by_css_selector('a')
#nextpage=nextpage.get_attribute('href')
#print(nextpage)
```
